```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Create a function that take in a dataframe, a quantitative_feature(or list), and a value for the multiplier.
# If you set extreme equal to True it will set the multiplier to 3.

def get_lower_and_upper_bounds(df, quantitative_features, multiplier=1.5, extreme=False):
    if extreme == True:
        # Create an empty dataframes that will eventually hold the observations that lie outside the upper and
        # lower limits.
        low_outliers = pd.DataFrame()
        high_outliers = pd.DataFrame()
        outliers = pd.DataFrame()
        # Set multiplier value to 3
        multiplier = 3
        # Create a for loop that will iterate through the list of quantitative values calculating the IQR value
        # and setting the upper and lower limits. 
        for feat in quantitative_features:
            q1 = df[feat].quantile(0.25)
            q3 = df[feat].quantile(0.75)
            # Calculate value for IQR
            iqr = q3 - q1
            # Calculate the upper and lower limits for acceptable values(inclusive)
            inner_lower_limit = q1 - (multiplier * iqr)
            inner_upper_limit = q3 + (multiplier * iqr)
            low_outliers = low_outliers.append(df[df[feat] < inner_lower_limit], ignore_index=True)
            print(f'Lower outliers for {feat}: {low_outliers}')
            high_outliers = high_outliers.append(df[df[feat] > inner_upper_limit], ignore_index=True)
            print(f'Upper outliers for {feat}: {high_outliers}')
            outliers = outliers.append(df[df[feat] < inner_lower_limit], ignore_index=True)
            outliers = outliers.append(df[df[feat] > inner_upper_limit], ignore_index=True)
        return low_outliers, high_outliers, outliers
    else:
        outliers = pd.DataFrame()
        # Create a for loop that will iterate through the list of quantitative values calculating the IQR value
        # and setting the upper and lower limits. 
        for feat in quantitative_features:
            q1 = df[feat].quantile(0.25)
            q3 = df[feat].quantile(0.75)
            # Calculate value for IQR
            iqr = q3 - q1
            # Calculate the upper and lower limits for acceptable values(inclusive)
            inner_lower_limit = q1 - (multiplier * iqr)
            inner_upper_limit = q3 + (multiplier * iqr)
            low_outliers = low_outliers.append(df[df[feat] < inner_lower_limit], ignore_index=True)
            print(f'Lower outliers for {feat}: {low_outliers}')
            high_outliers = high_outliers.append(df[df[feat] > inner_upper_limit], ignore_index=True)
            print(f'Upper outliersfor {feat}: {high_outliers}')
            outliers = outliers.append(df[df[feat] < inner_lower_limit], ignore_index=True)
            outliers = outliers.append(df[df[feat] > inner_upper_limit], ignore_index=True)
        return low_outliers, high_outliers, outliers
```
